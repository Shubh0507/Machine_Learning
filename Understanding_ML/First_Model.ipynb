1. Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

2. Import/Write the training set and the target set. 

```python

x_train = np.array([1, 2, 3, 4, 5])
y_train = np.array([3, 5, 7, 9, 11])

print(f"Training set: {x_train}")
print(f"Target set: {y_train}")
```

3. Calculate the number of elements in the training set

```python
m = x_train.shape[0]
print(f"Number of elements in the training set: {m}")
```

4. Plot the data to study

```python
plt.scatter(x_train, y_train, marker="o", c="r")
plt.title("Sample Regression Model")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

5. Set the values of w and b

```python
w = 2
b = 1
print(f"w: {w}")
print(f"b: {b}")
```

6. Create the output compute function

```python
def compute_model_output(x,w,b):
    m = x.shape[0]
    model_func_f = np.zeros(m)
    for i in range(0, m):
        model_func_f[i] = w * x[i] + b
    
    return model_func_f
```

7. Plot the output model against the given data

```python
f_wb = compute_model_output(x_train, w, b)

plt.plot(x_train, f_wb, c="b", label="Prediction by model")
plt.scatter(x_train, y_train, marker="o",c="k", label="Target values")
plt.title("Regression Model")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```
