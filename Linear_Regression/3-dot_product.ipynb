# DOT PRODUCT

Dot product calculates element by element multiplication in arrays much faster as compared to element by element multiplication using "For loop" as it uses the arrays as vectors. This is called vectorisation.

```python
import numpy as np 
```

```python
# Features: [size (1000 sqft), bedrooms]

X_train = np.array([
    [1.0, 1],
    [1.5, 2],
    [2.0, 2],
    [2.5, 3],
    [3.0, 3],
    [3.5, 4]
])

# Target: price (in $100k)
y = np.array([2.5, 3.5, 4.0, 5.0, 5.5, 6.5])


print(f"Training set: {X_train}")
print(f"Target set: {y}")
```

```python
# Precalculated values of b and w
w = np.array([1.0, 0.5]) # w is in array as it carries value for each feature
b = 1.0
print(f"w shape: {w.shape}, b type: {type(b)}")
```

## USING FOR LOOP

```python
def predict_for_loop(X, w, b):
    n = X.shape[0] # Finds the number of elements in the row
    p = 0 # initial cost = 0
    for i in range(n):
        p += w[i] * X[i]
    p += b

    return p
```

```python
x_vec = X_train[0,:]
f_wb = predict_for_loop()
```

## USING VECTORISATION

```python
def predict_vec(X, w, b): # This function is used to predict the value of target variable vased on a single row (acts as f_wb for the ith row)
    pred_val = np.dot(X, w) + b
    return pred_val
```

```python
# Now, given our data set, we know that for the 0th row, the target value is 2.5
x_vec = X_train[0,:]
f_wb = predict_vec(x_vec, w, b)

print(f"predicted value: {f_wb}")
```

Thus, we can see that we can make predictions and calculatation of the cost function faster using vectorisation.
