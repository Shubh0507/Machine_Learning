# FEATURE SCALING

Feature scaling is performed when the features are in different ranges or have a very large gap between them. This might result in divergence during the GDA. We have three main ways to perform feature scaling:- 
1. Division by max value
2. Mean normalization
3. z-score normalization (Most optimal)

### Z-SCORE NORMALIZATION

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/houses.csv")
df.head()
```

```python
df.columns = df.columns.str.strip()

X_train = df.drop("Price(1000s dollars)", axis=1).values 
y = df["Price(1000s dollars)"].values
```

```python
def z_score_normalisation(X):
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)

    X_norm = (X - mu) / sigma

    return X_norm, mu, sigma
```

```python
X_norm, X_mu, X_sigma = z_score_normalisation(X_train)
```
